### Supprt Vector Machines

```python
# import the libriaries
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
iris_data=datasets.load_iris() #load iris data from dataset
X=iris_data["data"][:,(2,3)]  # Petal Length and petal Width
y=(iris_data["target"][:].astype(np.float)) #Iris-Virginica (float values)
```

```python
#Linear SVM
pipe_svm=Pipeline([("scaler",StandardScaler()),
                   ("linear_svc",LinearSVC(C=1,loss="hinge")),])
pipe_svm.fit(X,y)
```

```python
pipe_svm.predict([[1.4,1.7]])
```
